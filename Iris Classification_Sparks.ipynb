## 1.Loading Libraries

```python
import pandas as pd                                     # import pandas for data structure support
from sklearn import datasets                            # import inbuilt datasets from scikit learn
from sklearn.model_selection import train_test_split    # use scikit learn for hold out
from sklearn import tree                                # import Decision Tree Classifier
from sklearn import metrics                             # import performance metrics
import matplotlib.pyplot as plt
```

## 2.Loading Dataset Iris

```python
# Loading load_iris() dataset

iris = datasets.load_iris() 
iris
```

```python
iris1 = pd.DataFrame(iris.data , columns= iris.feature_names)
iris1["Species"] = iris.target
iris1.head(5)
```

```python
iris1.shape
```

```python
# how many datapoints are there in each class
iris1["Species"].value_counts()
```

##  3.Creating Hold-Out Environment

```python
# Creating hold-out environmnet

irisdata_train, irisdata_test, iristarget_train, iristarget_test = train_test_split(iris.data, iris.target, test_size=0.3)
```

## 4.Building Decision Tree Model

```python
# Create a Decision tree classifier

DT_created = tree.DecisionTreeClassifier()

# Train the model with using training sets

DT_trained = DT_created.fit(irisdata_train, iristarget_train)
```

## 5.Testing the trained Decision Tree on the test data

```python
#Predict the response on the test data set

DT_predictions_1 = DT_trained.predict((irisdata_test))
```

## 6. Finding the accuracy of the trained model

```python
# Computing the acurracy of the Model

print("Accuracy:", round(metrics.accuracy_score(iristarget_test, DT_predictions_1),2) *100, "%")
```
